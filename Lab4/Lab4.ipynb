```python
import numpy as np
from math import copysign

# A function that evaluates the polynom we are trying to
# find roots from

def polynom(x):
    r = x**5 - 59 * x**4 + 35 * x**3 - 250 * x**2 + x - 70
    return(r)
```

```python
# we create a generator from tuples from the value and 
# the evaluation of the polynom using a 0.000001 precision

evaluated_polynom = ((x, polynom(x)) for x in np.arange(-1000, 1000, 0.000001))

# we get the min value from the absolute of the evaluations

answer = min(evaluated_polynom, key= lambda x: abs(x[1]))

# we print the answer

print(answer)
```

```python
# We intialize values of reference
# the limits of our evaluation
# the actual answer and a previous 
# answer that is used for check if it converges
# and the number of partitions we gonna use

limits = (-1000, 1000)
answer = np.mean(limits)
partitions = 3
previous_answer = np.nan

# we check if the answer equal the previous answer for check convergence

while answer != previous_answer:

    # we make ur partitions
    # get the signs of the evaluations

    limit_candidates = np.linspace(*limits, partitions) # type: ignore
    candidates_sign = tuple((copysign(1, polynom(x))
                             for x in limit_candidates))

    # start to check for every limit sign ans his next

    for pos, value in enumerate(candidates_sign[:-1]):

        # if the sign are different we have a ) inside the interval

        if value != candidates_sign[pos + 1]:

            # we save the answer as the previous
            # we get the new limits for the partitions
            # we get the new answer candidate
            # and we restart the number of partitions to the basic 3
            previous_answer = answer
            limits = (limit_candidates[pos], limit_candidates[pos + 1])
            answer = np.mean(limits)
            partitions = 3
            break

    # if we dont find any sign change in the partitions
    # then we redo the process with same limits but more partitions

    else:
        partitions +=1

# when the partitions get so small that the answers are the
# same twice due to computational number limit we found our answer
# and print it with his evaluation

print(answer, polynom(answer))

```
