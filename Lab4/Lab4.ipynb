```python
import numpy as np
from functools import partial
from math import copysign

def polynom(x):
    r = x**5 - 59 * x**4 + 35 * x**3 - 250 * x**2 + x - 70
    return(r)
```

```python
evaluated_polynom = ((x, polynom(x)) for x in np.arange(-1000, 1000, 0.00001))
answer = min(evaluated_polynom, key= lambda x: abs(x[1]))
print(answer)
```

```python
limits = (-1000, 1000)
answer = np.mean(limits)
partitions = 3
previous_answer = np.nan

while answer != previous_answer:
    limit_candidates = np.linspace(*limits, partitions) # type: ignore
    candidates_sign = tuple((copysign(1, polynom(x))
                             for x in limit_candidates))
    for pos, value in enumerate(candidates_sign[:-1]):
        if value != candidates_sign[pos + 1]:
            previous_answer = answer
            limits = (limit_candidates[pos], limit_candidates[pos + 1])
            answer = np.mean(limits)
            partitions = 3
            break
    else:
        partitions +=1

print(answer, polynom(answer))

```
